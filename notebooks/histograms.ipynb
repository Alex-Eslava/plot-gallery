## Fixed Number of Bins Histogram

```python
import numpy as np
import random
import math
from matplotlib import pyplot as plt

data = np.random.normal(0, 20, 1000) 
n_bins = 20
bins = np.linspace(math.ceil(min(data)), 
                   math.floor(max(data)),
                   n_bins) # fixed number of bins

plt.xlim([min(data)-5, max(data)+5])
plt.hist(data, bins=bins, alpha=0.5, ec='black')
plt.title('Fixed Number of Bins Histogram')
plt.xlabel('x')
plt.ylabel('count')

plt.show()

```

## Fixed Bin Size Histogram

```python
import numpy as np
import random
from matplotlib import pyplot as plt

data = np.random.normal(0,20,1000)
bin_size=5
bins = np.arange(-100, 100, bin_size)


plt.xlim([min(data)-5, max(data)+5])
plt.hist(data, bins=bins, alpha=0.5, ec='black')

plt.title('Fixed Bin Size Histogram')
plt.xlabel('x')
plt.ylabel('count')

plt.show()
```

## Overlapping Histograms

```python
import numpy as np
import random
from matplotlib import pyplot as plt

data1 = [random.gauss(15,10) for i in range(500)]  
data2 = [random.gauss(0,10) for i in range(500)]  
bin_size = 2.5
bins = np.arange(-60, 60, bin_size)

plt.xlim([min(data1+data2)-5, max(data1+data2)+5])

plt.hist(data1, bins=bins, alpha=0.3, label='class A')
plt.hist(data2, bins=bins, alpha=0.3, label='class B')
plt.title('Overlapping Histograms')
plt.xlabel('x')
plt.ylabel('count')
plt.legend(loc='upper right')


plt.show()
```
