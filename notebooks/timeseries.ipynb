## Missing Entries Plot

```python
import pandas as pd
import numpy as np

date_rng = pd.date_range(start='1/1/2020', end='1/1/2021', freq='D')
df = pd.DataFrame(date_rng, columns=['Date'])
df['Sales'] = np.random.randint(0, 365, size=(len(date_rng)))
df.loc[np.random.choice(df.index, 5), 'Sales'] = np.nan
df.set_index('Date', inplace=True)

start_date = '2020-04-04'
end_date = '2020-05-04'
df.loc[start_date:end_date, 'Sales'] = np.nan
```

```python
import matplotlib.pyplot as plt
import missingno as msno
# Can set the axis to be dates if I do a pivot but you get the point
msno.matrix(df, fontsize=20)
plt.show()
```
